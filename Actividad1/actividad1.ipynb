```python
# Librerías
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_curve,
    auc,
    precision_recall_curve,
    average_precision_score
)

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
np.random.seed(42)

```

```python
df = pd.read_csv('data-churn.csv')
df.head()
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
print("Tamaño del dataset:", df.shape)

print(df["Churn"].value_counts(normalize=True).rename("proporción"))

df.describe(include="all").transpose().head(20)

```

Se realizó una exploración inicial del dataset, analizando su tamaño, la distribución de la variable objetivo y estadísticas descriptivas de todas las variables. Se observa un desbalance de clases en la variable Churn, con una proporción significativamente mayor de clientes que no se fugan, lo cual motiva el uso de métricas como Precision, Recall, F1 y PR-AUC en la evaluación de los modelos.

```python
# Se definen variables numéricas y categóricas con las que se va a trabajar el modelo. 
numeric_features = [
    "MonthlyCharges",
    "TotalCharges",
    "tenure"
]

categorical_features = [
    "Contract",
    "InternetService",
    "PaymentMethod",
    "PhoneService"
]

X = df[numeric_features + categorical_features]
y = df["Churn"]
```

```python
# Se convierten las variables definidas como numéricas y se manejan datos que puedan arrojar errores
for col in numeric_features:
    X[col] = pd.to_numeric(X[col], errors="coerce")

```

Las variables numéricas fueron convertidas explícitamente a tipo numérico utilizando pd.to_numeric con errors="coerce", de modo que valores no numéricos fueran tratados como valores faltantes (NaN). Esta estrategia permite una detección e imputación consistente de datos faltantes y evita errores en etapas posteriores del preprocesamiento y entrenamiento del modelo.

```python
# Se eliminan los clientes con datos faltantes y se modifican los índices
X = X.dropna()
y = y[X.index]
```

```python
# Escalar numéricas
numeric_transformer_base = Pipeline(steps=[
    ("scaler", StandardScaler())
])

#Codificar categorías
categorical_transformer = Pipeline(steps=[
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Unir todo correctamente
preprocessor_base = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_base, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

Tras la conversión de variables numéricas, se eliminaron las filas con valores faltantes y se realineó el vector objetivo para mantener la correspondencia entre observaciones. Posteriormente, se definió un flujo de preprocesamiento utilizando ColumnTransformer, donde las variables numéricas fueron estandarizadas mediante StandardScaler y las variables categóricas codificadas mediante One-Hot Encoding. Este preprocesamiento se integró en un pipeline para asegurar una correcta aplicación durante la validación cruzada.

```python
def evaluate_model_cv(model, X, y, cv_splits=5):
    """
    Ejecuta k-fold CV estratificada y:
    - Ajusta el modelo en cada fold
    - Acumula predicciones y probabilidades
    - Muestra matriz de confusión, reporte, ROC, PR
    """
    skf = StratifiedKFold(n_splits=cv_splits, shuffle=True, random_state=42)

    y_true_all = []
    y_pred_all = []
    y_proba_all = []

    for train_idx, test_idx in skf.split(X, y):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]

        y_true_all.extend(y_test)
        y_pred_all.extend(y_pred)
        y_proba_all.extend(y_proba)

    y_true_all = np.array(y_true_all)
    y_pred_all = np.array(y_pred_all)
    y_proba_all = np.array(y_proba_all)

    # Matriz de confusión
    cm = confusion_matrix(y_true_all, y_pred_all)
    print("Matriz de confusión global (todos los folds):")
    print(cm)

    # Reporte de clasificación
    print("Reporte de clasificación:")
    print(classification_report(y_true_all, y_pred_all, digits=3))

    # ROC
    fpr, tpr, _ = roc_curve(y_true_all, y_proba_all)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC (AUC = {roc_auc:.3f})")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate (Recall)")
    plt.title("Curva ROC (CV global)")
    plt.legend()
    plt.show()

    # Precision–Recall
    precision, recall, _ = precision_recall_curve(y_true_all, y_proba_all)
    pr_auc = average_precision_score(y_true_all, y_proba_all)

    plt.figure()
    plt.step(recall, precision, where="post", label=f"PR (AP = {pr_auc:.3f})")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Curva Precision–Recall (CV global)")
    plt.legend()
    plt.show()

    print(f"AUC-ROC global: {roc_auc:.3f}")
    print(f"PR-AUC (Average Precision) global: {pr_auc:.3f}")

    return {
        "cm": cm,
        "roc_auc": roc_auc,
        "pr_auc": pr_auc
    }
```

La evaluación de los modelos se realizó mediante validación cruzada estratificada de 5 folds. Para cada partición, se entrenó el modelo y se obtuvieron predicciones de clase y probabilidades. Posteriormente, se acumularon las predicciones de todos los folds para construir métricas globales, incluyendo matriz de confusión, curvas ROC y Precision–Recall. Este enfoque permite una evaluación más estable y facilita la comparación entre distintos modelos.

```python
log_reg_base = Pipeline(steps=[
    ("preprocess", preprocessor_base),
    ("clf", LogisticRegression(
        penalty="l2",   # C grande ≈ sin regularización fuerte
        C=1e6,
        solver="lbfgs",
        max_iter=1000
    ))
])


results_base = evaluate_model_cv(log_reg_base, X, y, cv_splits=5)
```

El modelo base de regresión logística presenta una buena capacidad discriminatoria, con un AUC-ROC de 0.839. Sin embargo, al analizar la matriz de confusión y las métricas de la clase churn, se observa un recall de 0.526, lo que indica que una proporción significativa de clientes que efectivamente se fugan no es detectada por el modelo. Aunque la accuracy global es cercana al 80%, esta métrica resulta engañosa debido al desbalance de clases. En este contexto, métricas como F1 y PR-AUC resultan más informativas para evaluar el desempeño real del modelo desde una perspectiva de negocio.

La curva Precision–Recall muestra claramente el trade-off entre detectar una mayor proporción de clientes que se fugan (recall) y mantener una precisión razonable. En el punto de operación del modelo base, el recall es cercano a 0.53 con una precision de aproximadamente 0.64. Al aumentar el recall, la precision disminuye de forma significativa, lo que implica un mayor número de falsos positivos. Dado el desbalance del problema, la métrica PR-AUC (0.633) resulta más informativa que la accuracy para evaluar el desempeño del modelo.

```python
numeric_transformer_poly = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("poly", PolynomialFeatures(degree=4, include_bias=False))
])

preprocessor_poly = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_poly, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

log_reg_poly = Pipeline(steps=[
    ("preprocess", preprocessor_poly),
    ("clf", LogisticRegression(
        penalty="l2",
        C=1e6,
        solver="lbfgs",
        max_iter=1000
    ))
])

results_poly = evaluate_model_cv(log_reg_poly, X, y, cv_splits=5)
```

Al incorporar transformaciones polinomiales de grado 4 sobre las variables numéricas, el modelo incrementa significativamente su complejidad. Los resultados muestran una leve mejora en métricas globales como ROC-AUC (0.841) y PR-AUC (0.643), однако el recall de la clase churn disminuye respecto al modelo base, pasando de 0.526 a 0.479. Este comportamiento indica que, si bien el modelo discrimina ligeramente mejor, pierde capacidad para detectar clientes que efectivamente se fugan, evidenciando un posible sobreajuste. Por esta razón, se hace necesario aplicar regularización para controlar la complejidad del modelo.

```python
base_pipeline_poly = Pipeline(steps=[
    ("preprocess", preprocessor_poly),
    ("clf", LogisticRegression(max_iter=1000))
])
```

```python
param_grid_l2 = {
    "clf__penalty": ["l2"],
    "clf__solver": ["lbfgs"],
    "clf__C": [0.01, 0.1, 1.0, 10.0]
}

grid_l2 = GridSearchCV(
    base_pipeline_poly,
    param_grid_l2,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring="average_precision",  # PR-AUC
    n_jobs=-1
)

grid_l2.fit(X, y)

print("Mejores hiperparámetros (L2):", grid_l2.best_params_)
print("Mejor PR-AUC medio (L2):", grid_l2.best_score_)

best_model_l2 = grid_l2.best_estimator_
```

Para controlar la complejidad introducida por las transformaciones polinomiales, se aplicó regularización L2 mediante una búsqueda en grilla de hiperparámetros. Utilizando PR-AUC como métrica de optimización, se observó que el mejor desempeño se obtiene con un valor intermedio de la fuerza de regularización (C = 0.1), alcanzando un PR-AUC promedio de 0.647. Estos resultados indican que la regularización L2 permite mejorar la capacidad de generalización del modelo polinomial y mitigar el sobreajuste observado previamente.

```python
param_grid_l1 = {
    "clf__penalty": ["l1"],
    "clf__solver": ["liblinear"],
    "clf__C": [0.01, 0.1, 1.0, 10.0]
}

grid_l1 = GridSearchCV(
    base_pipeline_poly,
    param_grid_l1,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring="average_precision",
    n_jobs=-1
)

grid_l1.fit(X, y)

print("Mejores hiperparámetros (L1):", grid_l1.best_params_)
print("Mejor PR-AUC medio (L1):", grid_l1.best_score_)

best_model_l1 = grid_l1.best_estimator_
```

Se evaluó además una regresión logística con penalización L1, la cual promueve modelos más esparsos al inducir coeficientes nulos. Mediante Grid Search con validación cruzada estratificada y optimización de PR-AUC, se obtuvo un mejor desempeño con C = 1.0, alcanzando un PR-AUC promedio de 0.646. El rendimiento es comparable al obtenido con regularización L2, aunque L1 ofrece la ventaja adicional de mayor interpretabilidad al reducir el número de variables activas en el modelo.

```python
results_l2 = evaluate_model_cv(best_model_l2, X, y, cv_splits=5)

results_l1 = evaluate_model_cv(best_model_l1, X, y, cv_splits=5)
```

## Desición

Se selecciona el modelo con regularización L2 dado que presenta el mayor PR-AUC promedio, métrica prioritaria en un problema desbalanceado como churn, además de un desempeño estable en todas las métricas evaluadas.


## Conclusión

Finalmente, se evaluaron los modelos polinomiales regularizados con L1 y L2 utilizando validación cruzada estratificada. Ambos enfoques presentan desempeños muy similares, con PR-AUC cercanos a 0.65 y diferencias marginales en precision y recall para la clase churn. La regularización L2 obtiene el mayor PR-AUC, mientras que L1 logra un recall ligeramente superior. Dado el carácter desbalanceado del problema y la importancia de detectar clientes en riesgo, ambas alternativas resultan válidas, evidenciando que la regularización es fundamental para controlar el sobreajuste introducido por las transformaciones polinomiales.
